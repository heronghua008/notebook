## epoll
- 使用方法：与poll基本相同
  - 生成对象使用epoll() 而不是poll()
  - 将register注册IO事件时 关注的事件类别改为epoll类别

## select  poll   epoll 区别
- select 可以很好支持windows
- epoll 比 select 和 poll 效率高， select 和 poll 差不多
- epoll提供了更多的触发方式

## 本地套接字

- linux下文件类型
  - b（块设备文件）
  - c（字符设备文件）
  - d（目录）
  - \-（普通文件）
  - l（链接文件）
  - s（套接字文件）
