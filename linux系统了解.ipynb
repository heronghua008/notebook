# Linux 的使用：
    Linux/Unix 命令
      命令名  [选项]  [参数]
      注：[]的内容代表可选
      
    命令示例：
      ls         # 显示当前文件夹下的所有文件和文件夹
      ls - s     # 列表显示所有文件
      pwd        # 显示当前的路径
      cd /       # 切换到根目录
      cd         # 切换到当前用户主目录
      
# Linux 常用命令：
    clear 命令：
      作用： 清屏  快捷键 ctrl + l
    pwd 命令：
      作用： 用于显示当前工作路径的位置
      示例：
         heronghua@mark:~￥ pwd
         /home/heronghua
    ls 命令：
        作用：显示文件夹内容
        格式: ls [选项] [文件夹名称或文件名]
        常用选项：
            -l 列表显示文件的详细信息
            -a 显示全部文件/文件夹
    cd 命令:
        作用：改变当前工作路径（进入某个目录）
        格式： cd[路径] 
    mkdir 命令：
        作用：创建一个或多个文件夹
        格式：mkdir 文件夹名1 文件夹名2 。。。
        示例：
            mkdir a b c d
            mkdir -p a/bb/ccc/ddd
        常用选项：
            -p 如果中间的文件夹不存在，则逐级创建文件夹
    rmdir 命令：
        作用：删除一个或多个文件夹（文件夹内必须为空）
        格式：rmdir [选项] 文件夹名
        常用选项：
            -p 如果中间文件夹也不空，则逐级删除中间文件夹
         示例：
             rmdir a b c d
             rmdir -p a/bb/ccc/ddd
    tree 命令：
        作用：
            显示目录树：
        示例：
            tree
            tree day01
    touch 命令：
        作用：
            1.创建新的文件
            2.如果文件存在，则更新文件的修改时间
        格式：
            touch 文件名
    rm 命令：
        作用：删除文件或者文件夹
        格式：rm [选项] 文件/文件夹
        示例：
            rm a.txt
            rm a.txt b.txt
            rm *.txt
            rm -r a
        常用选项：
            -r 递归删除文件夹内的文件和文件夹
            -i 删除前给出提示（y代表yes,n代表no）
            -f 强制删除，不给任何提示
    操作带有特殊符号的路径的方法：
        用两个双引号（""）把路径括起来
        例：
            mkdir "a b"
            rmdir "a b"
    help 选项：
        作用：查看此命令的帮助信息
        格式：命令名 --help
        示例：
            ls --help
            rm --help
    man 命名帮助：
        格式： man Linux/Unix命令名
        作用： 查看命令所对应的帮助文档（手册）
        示例：
            man ls
            man rm
           翻页： 上下健
           退出： 按q退出
    快捷键：
        Tab 补全命令和路径
        上下键 翻出以前的操作过的命令
       
# Linux/Unix 下的路径：
     路径：
         是用来记录一个文件或文件夹的字符串
         如：
             /home/tarena
             /home/tarena/a.txt
     路径分为两种：
         绝对路径：
             以‘/’字符开头的路径为绝对路径（在任何时候，一个文件的绝对路径都是唯一的）
             
         相对路径：
             1) 文件/文件夹名称
             2) .当前的文件夹
             3）..上一级文件夹
             4）～ 用户主目录（家目录）
